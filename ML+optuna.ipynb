```python
def all_models(X_train, X_test, y_train, y_test):
    models = {
        'random_forest': RandomForestClassifier(),
        'Decision_tree': DecisionTreeClassifier()
    }

    # Parameter grids for GridSearchCV
    param_grid = {
        'random_forest': {
            'n_estimators': [50, 100],
            'max_depth': [None, 10],
            'min_samples_split': [2, 5]
        },
        'Decision_tree': {
            'max_depth': [None, 10],
            'min_samples_split': [2, 5]
        }
    }
    
    scores =[]
    for name, models in models.items():

        grid_search = GridSearchCV(models, param_grid[name], cv=5, scoring='accuracy')

        grid_search.fit(X_train, y_train)
        y_pred =grid_search.predict(X_test)

        acc = accuracy_score(y_test, y_pred)
        scores.append((name, acc))
        a = print(grid_search.best_params_)
        df = pd.DataFrame(scores, columns=['name', 'accuracy'])  
    return df

```

```python
all_models(X_train, X_test, y_train, y_test)
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('tip', axis = 1)
y = df['tip']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
tr1 = ColumnTransformer(
    transformers = [
        ('OHE', OneHotEncoder(drop='first', sparse_output=True), ['sex', 'time', 'smoker']),
        ('OE', OrdinalEncoder(categories=[['Sun', 'Sat', 'Thur', 'Fri']]), ['day']),
        ('ST', StandardScaler(), ['total_bill'])
    ],
    remainder='passthrough'
)

Scaled_X_train = tr1.fit_transform(X_train)
scaled_X_test = tr1.transform(X_test)
```

```python
from flaml import AutoML
```

```python
# Initialize and fit the AutoML model
automl = AutoML()

auto_ml_settings = {
    'time_budget': 60,
    'metric': 'r2',
    'task': 'classification',
    'estimator_list': [ 'rf', 'lgbm', 'lgbm_spark', 'lrl1', 'lrl2', 'catboost', 'extra_tree', 'kneighbor',  'histgb']

}

automl.fit(Scaled_X_train, y_train, **auto_ml_settings)

# Evaluate the model
y_pred = automl.predict(scaled_X_test)
```

```python

print(f'MAE: {mean_absolute_error(y_test, y_pred)}')
```
